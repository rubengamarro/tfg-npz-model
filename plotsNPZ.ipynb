#Gráfica comparación entre la tasa de pastoreo de Ivlev y la de Mayzaud-Poulet

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros
Rm = 1  # Tasa máxima de ingesta
Lambda = 0.5 # Coeficiente de eficiencia

# Rango de P (concentración de presas)
P = np.linspace(0, 5, 100)
```

```python
# Definición de las funciones
I1 = Rm * (1 - np.exp(-Lambda * P))
I2 = Rm * Lambda * P * (1 - np.exp(-Lambda * P))
```

```python
# Gráficas
plt.figure(figsize=(8, 5))

plt.plot(P, I1, label=r'$h_1(P) = R_m \left( 1-e^{-\Lambda P} \right)$ → Ivlev', color='b')
plt.plot(P, I2, label=r'$h_2(P) = R_m \Lambda P \left( 1-e^{-\Lambda P} \right)$ → Mayzaud y Poulet', color='r', linestyle='--')

plt.xlabel('P (Concentración de fitoplancton)')
plt.ylabel('Tasa de pastoreo (por día)')
plt.title('Curvas de Ingestión del Zooplancton')
plt.legend()
#plt.grid()

plt.show()
```
